5. Capture Hotel details

    2. description

```python
# Wait for the filter to apply
time.sleep(5)

# Scroll to load all hotel cards
SCROLL_PAUSE_TIME = 2

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to the bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load the page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with the last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
import pandas as pd
```

```python
div_cards = driver.find_elements(By.CSS_SELECTOR, '[data-testid="property-card"]')
```

```python
len(div_cards)
```

```python
hotel_links = []
```

```python
for i in tqdm(div_cards):
    hotel_links.append(i.find_element(By.TAG_NAME, 'a').get_attribute('href'))
```

```python
hotel_links
```

```python
from selenium.common.exceptions import StaleElementReferenceException, TimeoutException
```

```python
hotel_details = []
for link in tqdm(hotel_links):
    driver.get(link)
    try:
        hotel_name = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'h2.d2fee87262.pp-header__title'))
        ).text
    except:
        hotel_name = None
    
    try:
        hotel_description = WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, '[data-testid="property-description"]'))
        ).text
    except:
        hotel_name = None
        
#     print(hotel_name)
    
    hotel_details.append({
        "Hotel Name": hotel_name,
        "Description": hotel_description
    })
```

```python
df = pd.DataFrame(hotel_details)

df.to_csv('hotel_details.csv', index=False)
```
