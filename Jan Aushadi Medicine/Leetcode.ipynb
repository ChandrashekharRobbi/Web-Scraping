```python
browser.find_elements(By.XPATH, '//*[@id="__next"]/div/div[2]/div[1]/div[1]/div[5]/div[2]/div/div/div[2]/div[1]/div[2]/div/div/div/div/a')[0].text
```

```python
data = []
n = 0
page = 2
while n != 29:
    for i in range(1,101):
        title = browser.find_elements(By.XPATH, f'//*[@id="__next"]/div/div[2]/div[1]/div[1]/div[5]/div[2]/div/div/div[2]/div[{i}]/div[2]/div/div/div/div/a')[0].text
        link =  browser.find_elements(By.XPATH, f'//*[@id="__next"]/div/div[2]/div[1]/div[1]/div[5]/div[2]/div/div/div[2]/div[{i}]/div[2]/div/div/div/div/a')[0].get_attribute('href')
        acceptance = browser.find_element(By.XPATH, f'//*[@id="__next"]/div/div[2]/div[1]/div[1]/div[5]/div[2]/div/div/div[2]/div[{i}]/div[4]/span').text
        difficulty = browser.find_element(By.XPATH, f'//*[@id="__next"]/div/div[2]/div[1]/div[1]/div[5]/div[2]/div/div/div[2]/div[{i}]/div[5]/span').text
        if i == 100:
            browser.find_element(By.XPATH, f'//*[@id="__next"]/div/div[2]/div[1]/div[1]/div[5]/div[3]/nav/button[10]').click()
            time.sleep(4)
            page += 1
        data.append([title, acceptance, difficulty, link])
        print(title)
        print(acceptance)
        print(difficulty)
        print(link)
        print(f"n: {n}")
    n += 1
```

```python
df = pd.DataFrame(data, columns=["Questions", "Acceptance Rate", "Difficulty", "Links"])
```
