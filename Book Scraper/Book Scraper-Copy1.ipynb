### Extract all books and their prices from all pages

```python
import pandas as pd
from tqdm import tqdm
import requests
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up Selenium Chrome driver
chrome_options = Options()
chrome_options.add_argument('--headless')  # Run Chrome in headless mode (without GUI)
selenium_service = Service('../chromedriver_win32/')  # Replace 'path_to_chromedriver' with the actual path
driver = webdriver.Chrome(options=chrome_options)

# Navigate to the LeetCode problems page
url = 'https://leetcode.com/problemset/all/'
driver.get(url)

# Wait for the problem list to load
wait = WebDriverWait(driver, 10)
wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="__next"]/div/div[2]/div[1]/div[1]/div[6]/div[2]/div/div/div[2]')))

# Find the problem container
problem_container = driver.find_element(By.XPATH, '//*[@id="__next"]/div/div[2]/div[1]/div[1]/div[6]/div[2]/div/div/div[2]')

# Find all the problem elements within the container
problem_elements = problem_container.find_elements(By.CLASS_NAME, 'reactable-row')

# Iterate over each problem element
for element in problem_elements:
    # Extract problem name and link
    problem_name_elem = element.find_element(By.CLASS_NAME, 'text-gray')
    problem_name = problem_name_elem.text.strip()
    problem_link = problem_name_elem.get_attribute('href')

    # Extract acceptance rate
    acceptance_rate_elem = element.find_element(By.CLASS_NAME, 'text-success')
    acceptance_rate = acceptance_rate_elem.text.strip()

    # Extract difficulty level
    difficulty_elem = element.find_element(By.CLASS_NAME, 'difficulty-label')
    difficulty_level = difficulty_elem.text.strip()

    # Print the information
    print('Problem:', problem_name)
    print('Link:', problem_link)
    print('Acceptance Rate:', acceptance_rate)
    print('Difficulty Level:', difficulty_level)
    print('-' * 50)

# Quit the driver
driver.quit()
```

```python
data = []
for i in tqdm(range(1,51)):
    link = 'https://leetcode.com/problemset/all/?page='
    link1 = link + str(i) + '.html'  # it total has 50 pages
    res = requests.get(link1)
    soup = BeautifulSoup(res.text, 'html.parser')
    link2 = 'https://books.toscrape.com'
    link3 = 'https://books.toscrape.com/catalogue/'
    
    for i in soup.find_all('article',class_='product_pod'):
        # title name
        title = i.find_all('a')[-1].get('title')
        # href
        href_get = i.find_all('a')[-1].get('href')
        
        href = link3 + href_get
        # price
        price = i.find('p', class_='price_color').text[1:]
        
        # star rating classes
        star_class = i.find('p')['class']
        # star rating
        star_rating = star_class[1]
        
        # stock
        stock = i.find('p', class_='instock').text.strip()
        # image source
        image_src = i.find('img', class_='thumbnail')['src'][2:]
        
        
        # image link
        image_link = link2 + image_src
        data.append([title, href, price, star_rating, stock, image_link])
```
